기존보다 더 깊은 신경망으로 보다 복잡한 문제를 해결할 수 있는 딥러닝


## - 층을 깊게 하는 이유
### - 신경망의 매개 변수가 줄어든다



- 5x5의 매개변수는 25개 - 3x3은 18개로 매개 변수를 줄일 수 있다
- - 매개변수가 줄어든다는 점
  - 이는 연산에 들어갈 비용을 줄일 수 있다는 뜻이다


# 신경망 정확도를 높이기 위한 기술

- 데이터 확장
- - 인위적으로 이미지의 회전이나 이동 등 미세한 변화를 주어 이미지의 개수를 늘린다
